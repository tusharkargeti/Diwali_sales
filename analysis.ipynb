```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"Diwali Sales Data.csv",encoding='latin1')
df
```

```python
df.isnull().sum()
```

```python
df=df.drop(["Status","unnamed1"],axis=1)   # axis =1 is used for that we wnt drop column not row 
df
```

```python
# Calculate the mean of the 'Amount' column
mean_value = df["Amount"].mean()

# Fill NaN values with the mean value in place
df["Amount"].fillna(mean_value, inplace=True)

# df should now be correctly updated
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Amount']=df['Amount'].astype("int")
```

```python
df.info()
```

```python
df.rename(columns={"Marital_Status":"shadi"},inplace=True)
```

```python
df.info()
```

```python
#GENDER
```

```python
ax= sns.countplot(x="Gender",data=df,palette="Set1")    # pa
ax.set_title('Gender Distribution')
ax.set_xlabel('Gender')
ax.set_ylabel('Count')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_gen=df.groupby(['Gender'])['Amount'].sum().reset_index()
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen,palette="Set1")
```

```python
#  FROM ABOVE GRAPH WE CAN SEE THAT FEMALE PURCHASE MORE ITEM  AND ALSO SPEND MORE MONEY THAN MEN
```

```python
ax=sns.countplot(x="Age Group",data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'])['Amount'].sum().reset_index()
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age,palette="Set1")
```

```python
# FROM ABOVE GRAPH WE CAN SEE THAT FEMALE AND AGE GROUP OF 26-35 PURCHASE MORE 
```

```python
sales_state=df.groupby(['State'])['Orders'].sum().reset_index().sort_values(by='Orders',ascending=False).head(10)
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='State',y='Orders',data=sales_state,palette='Set1')

plt.show()
```

```python
# total/sales amount of state
sales_state = df.groupby(['State'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending=False).head(10)
```

```python
plt.figure(figsize=(25,10))
sns.barplot(x='State',y='Amount',data=sales_state,palette='Set1')

plt.show()
```

```python
# FROM ABOVE GRAPH WE CAN SEE THAT UTTARPRADESH ,MAHARSHTRA ,KARNATAKA ORDER THE MOST BUT RESPECTIVELy WE CAN ALSO SEE THAT KERALA ORDER 
# ARE MORE THAN HARyana BUT IN SALES AMOUNT HAryana are more than kerala
```
